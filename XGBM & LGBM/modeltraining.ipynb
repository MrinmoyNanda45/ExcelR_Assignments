### Light XBM

```python
import lightgbm as lgb
```

```python
def train_lightgbm_model(X_train,y_train):
    """
    Train a LightGBM model on the data.

    Parameters:
    - data (pd.DataFrame): The dataset containing the features and target.
    - target (str): The name of the target column.
    - categorical_features (list): List of categorical feature column names.

    Returns:
    - model (LGBMClassifier): The trained LightGBM model.
    - accuracy (float): The accuracy of the model on the test set.
    """
    lgb_model = lgb.LGBMClassifier()
    lgb_model.fit(X_train, y_train)
    
    return lgb_model
    
```

### XGBM

```python
from xgboost import XGBClassifier as xgb
```

```python
def train_xgboost_model(X_train, y_train):
    """
    Train an XGBoost model on the data.

    Parameters:
    - data (pd.DataFrame): The dataset containing the features and target.
    - target (str): The name of the target column.
    - categorical_features (list): List of categorical feature column names.

    Returns:
    - model (XGBClassifier): The trained XGBoost model.
    - accuracy (float): The accuracy of the model on the test set.
    """
    XGb_model = xgb(eval_metric='mlogloss',use_label_encoder=False)
    XGb_model.fit(X_train, y_train)
    
    return XGb_model
```

### Evaluation

### Evaluating Model on Titanic Dataset without Target Variable in Test Data

#### Why Can't We Use the Test Data for Evaluation?

In the **Titanic dataset** provided by Kaggle, the `titanic_test.csv` file only contains the feature columns (`X_test`) and lacks the target variable (`Survived`). The target variable is crucial for evaluating the performance of a classification model, as it allows us to compare the predicted values (`y_pred`) with the actual labels (`y_test`).

Since the `titanic_test.csv` file does not have the target variable (`y_test`), **we cannot directly evaluate the model's accuracy or performance using this dataset**. Evaluation metrics such as accuracy, precision, recall, or F1-score require the true labels (`y_test`) to compute how well the model has performed.

#### What Will We Use for Evaluation?

In order to evaluate the model's performance:

1. **Use the Titanic Train Dataset**: 
   - You will use the `titanic_train.csv` dataset to train the model. This dataset contains both features (`X_train`) and the target variable (`Survived`), which is used for training the model.
   
2. **Split the Data**:
   - If you do not have a separate validation set, you can split the `titanic_train.csv` dataset into a training set and a validation set (using `train_test_split` from scikit-learn).
   - This split will give you both the feature data (`X_train`) and the target variable (`y_train`), as well as validation data (`X_val` and `y_val`), which allows for performance evaluation.

---

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
def evaluate_model(model, X_test, y_test):
    """
    Evaluate the given model using various performance metrics.

    Parameters:
    - model: The trained XGBoost or LightGBM model.
    - X_test: The feature data for testing.
    - y_test: The true labels for testing.

    Returns:
    - None
    """
    # Predicting the labels using the trained model
    y_pred = model.predict(X_test)
    
    # Calculating evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    
    # Printing evaluation results
    print("Model Evaluation Results:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print("\nConfusion Matrix:")
    print(cm)
    print("\nClassification Report:")
    print(report)
```

### Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
def cross_validate_model(model, X, y, cv=5):
    """
    Perform cross-validation on the given model and return the performance scores.

    Parameters:
    - model: The trained model (XGBoost or LightGBM).
    - X: The feature data.
    - y: The target data.
    - cv: The number of cross-validation folds (default is 5).

    Returns:
    - mean_score: The average cross-validation score.
    - std_score: The standard deviation of the cross-validation scores.
    """
    scores = cross_val_score(model, X, y, cv=cv, scoring='accuracy')
    mean_score = np.mean(scores)
    std_score = np.std(scores)
    
    print(f"Cross-Validation Accuracy: {mean_score:.4f} Â± {std_score:.4f}")
    return mean_score, std_score
```

### Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
def perform_random_search(model, param_dist, X_train, y_train, n_iter=50, cv=5, scoring='accuracy', random_state=42, verbose=2):
    """
    Perform Randomized Search to tune the hyperparameters of the model.
    
    Args:
        model: The machine learning model to tune.
        param_dist (dict): The hyperparameter distribution.
        X_train (pd.DataFrame or np.ndarray): Training features.
        y_train (pd.Series or np.ndarray): Training labels.
        n_iter (int): Number of iterations for RandomizedSearchCV.
        cv (int): Number of cross-validation folds.
        scoring (str): Scoring metric for evaluation.
        random_state (int): Random state for reproducibility.
        verbose (int): Verbosity level.
        
    Returns:
        RandomizedSearchCV: Trained RandomizedSearchCV object.
    """
    random_search = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_dist,
        n_iter=n_iter,
        cv=cv,
        scoring=scoring,
        verbose=verbose,
        random_state=random_state,
        n_jobs=-1  # Use all available processors
    )
    random_search.fit(X_train, y_train)
    return random_search.best_estimator_ , random_search.best_params_
```

### XGBM

```python
xgb_param_dist = {
    'n_estimators': [100, 200, 300, 400],
    'learning_rate': np.linspace(0.01, 0.3, 10),
    'max_depth': [3, 5, 7, 9],
    'min_child_weight': [1, 3, 5, 7],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0],
    'gamma': [0, 0.1, 0.2, 0.3],
    'reg_alpha': np.logspace(-3, 1, 5),
    'reg_lambda': np.logspace(-3, 1, 5)
}
```

```python
best_model_XGBM, best_params_XGBM = perform_random_search(model= XGB_model,
                                                                     param_dist= xgb_param_dist,
                                                                     X_train= X_train,
                                                                     y_train= y_train)
```
