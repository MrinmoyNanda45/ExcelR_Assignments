## Spliting the data into independent and dependent

```python
X = df.drop('Price',axis=1)
y = df['Price']
```

## Spliting the data as training and test sets

```python
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test) #Just use tranform on test data to avoid data leakage
```

## Prediction

## Lasso Regularization

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
```

```python
model = Lasso()
```

```python
parameter = {'alpha' : np.logspace(-4, 0, 10)}
```

```python
grid = GridSearchCV(model,parameter,cv= 5,scoring= 'neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
# Best alpha value found
best_alpha = grid.best_params_['alpha']
print(f"Best alpha: {best_alpha}")
```

```python
# Train a Lasso model using the best alpha value
best_lasso = Lasso(alpha=best_alpha)
best_lasso.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = best_lasso.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
r2 = r2_score(y_test, y_pred)  # R-squared score
```

```python
# Display the results
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
print('Coefficients:', best_lasso.coef_)
print('Intercept:', best_lasso.intercept_)
```

## Ridge

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge()
```

```python
# Define a range of alpha values for hyperparameter tuning
alpha_range = {'alpha': np.logspace(-4, 2, 10)}  # From 0.0001 to 100
```

```python
# Use GridSearchCV to find the best alpha
grid = GridSearchCV(estimator=model, param_grid=alpha_range, cv=5, scoring='neg_mean_squared_error')
```

```python
# Fit the grid search to the training data
grid.fit(X_train, y_train)
```

```python
# Best alpha value found
best_alpha = grid.best_params_['alpha']
print(f"Best alpha: {best_alpha}")
```

```python
# Train a Ridge model using the best alpha value
best_ridge = Ridge(alpha=best_alpha)
best_ridge.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = best_ridge.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
r2 = r2_score(y_test, y_pred)  # R-squared score
```

```python
# Display the results
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
print('Coefficients:', best_ridge.coef_)
print('Intercept:', best_ridge.intercept_)
```

## Elastic Net

```python
from sklearn.linear_model import ElasticNet
```

```python
model = ElasticNet()
```

```python
# Define a grid of alpha and l1_ratio values for hyperparameter tuning
param_grid = {
    'alpha': np.logspace(-4, 0, 10),  # From 0.0001 to 1
    'l1_ratio': np.linspace(0, 1, 10)  # From 0 (Ridge) to 1 (Lasso)
}
```

```python
# Use GridSearchCV to find the best combination of alpha and l1_ratio
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
```

```python
# Fit the grid search to the training data
grid_search.fit(X_train, y_train)
```

```python
# Best alpha and l1_ratio values found
best_alpha = grid_search.best_params_['alpha']
best_l1_ratio = grid_search.best_params_['l1_ratio']
print(f"Best alpha: {best_alpha}")
print(f"Best l1_ratio: {best_l1_ratio}")
```

```python
# Train an Elastic Net model using the best alpha and l1_ratio values
best_elastic_net = ElasticNet(alpha=best_alpha, l1_ratio=best_l1_ratio)
best_elastic_net.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = best_elastic_net.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
r2 = r2_score(y_test, y_pred)  # R-squared score
```

```python
# Display the results
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
print('Coefficients:', best_elastic_net.coef_)
print('Intercept:', best_elastic_net.intercept_)
```
