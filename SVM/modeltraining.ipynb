### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
def scale_data(X_train, X_test):
    """
    Scales the training and test datasets using StandardScaler.
    
    Args:
        X_train (pd.DataFrame or np.array): Training feature data
        X_test (pd.DataFrame or np.array): Test feature data
        
    Returns:
        X_train_scaled (pd.DataFrame or np.array): Scaled training data
        X_test_scaled (pd.DataFrame or np.array): Scaled test data
    """
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    return X_train_scaled, X_test_scaled
```

```python
X_train_scaled, X_test_scaled = scale_data(X_train= X_train, X_test= X_test)
```

### Model Training

```python
from sklearn.svm import SVC
```

```python
def train_svm_classifier(X_train, y_train):
    """
    Trains an SVM classifier with the training data.
    
    Args:
        X_train (pd.DataFrame or np.array): Scaled training feature data
        y_train (pd.Series or np.array): Training target data
        
    Returns:
        model (SVC): Trained SVM classifier model
    """
    model = SVC() 
    model.fit(X_train, y_train)
    
    return model
```

### Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
def evaluate_model(model, X_test, y_test):
    """
    Evaluates the trained model on the test dataset.
    
    Args:
        model (SVC): Trained SVM classifier model
        X_test (pd.DataFrame or np.array): Scaled test feature data
        y_test (pd.Series or np.array): Test target data
        
    Returns:
        None: Prints evaluation metrics
    """
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='binary', pos_label='poisonous')
    recall = recall_score(y_test, y_pred, average='binary', pos_label='poisonous')
    f1 = f1_score(y_test, y_pred, average='binary', pos_label='poisonous')
    
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    
    
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
evaluate_model(model= SVC_model, X_test= X_test, y_test= y_test)
```

### Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint
```

```python
def svm_random_search(X_train, y_train, X_test, y_test, n_iter=100, cv=5, n_jobs=-1, random_state=42):
    """
    This function performs hyperparameter tuning for an SVM model using RandomizedSearchCV,
    trains the best model, and evaluates it on the test data.
    
    Parameters:
    - X_train: Training features
    - y_train: Training labels
    - X_test: Test features
    - y_test: Test labels
    - n_iter: Number of iterations for RandomizedSearchCV (default is 100)
    - cv: Number of cross-validation folds (default is 5)
    - n_jobs: Number of parallel jobs for RandomizedSearchCV (default is -1, meaning all processors)
    - random_state: Random seed for reproducibility (default is 42)
    
    Returns:
    - best_model: The best model found from RandomizedSearchCV
    - best_params: The best hyperparameters found
    - report: The classification report on the test data
    """
    param_dist = {
        'C': uniform(0.1, 10),
        'kernel': ['linear', 'rbf', 'poly'],
        'gamma': ['scale', 'auto'],
        'degree': randint(2, 6)
    }
    svm = SVC()
    random_search = RandomizedSearchCV(estimator=svm, param_distributions=param_dist, 
                                       n_iter=n_iter, cv=cv, n_jobs=n_jobs, 
                                       verbose=2, random_state=random_state, scoring='accuracy')

    random_search.fit(X_train, y_train)

    best_model = random_search.best_estimator_
    best_params = random_search.best_params_

    y_pred = best_model.predict(X_test)
    report = classification_report(y_test, y_pred)

    print(f"Best hyperparameters found: {best_params}")
    print("\nClassification Report:\n", report)

    return best_model, best_params, report
```

```python
best_model, best_params, report = svm_random_search(X_train= X_train, X_test= X_test, y_train= y_train, y_test= y_test)
```

### Visualization

```python
from sklearn.metrics import confusion_matrix
```

```python
def visualize_classification_results(model, X_test, y_test, class_names):
    """
    Visualizes classification results by plotting the decision boundaries and 
    true vs predicted labels on the test data, along with a confusion matrix.
    
    Parameters:
    - model: Trained model (SVM in this case)
    - X_test: Test features
    - y_test: True labels of the test data
    - class_names: List of class names (labels)
    """
    # Predict on the test data
    y_pred = model.predict(X_test)
    
    # Plotting the decision boundaries (for 2D data)
    if X_test.shape[1] == 2:
        h = .02  # Step size in the mesh
        x_min, x_max = X_test[:, 0].min() - 1, X_test[:, 0].max() + 1
        y_min, y_max = X_test[:, 1].min() - 1, X_test[:, 1].max() + 1
        
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                             np.arange(y_min, y_max, h))
        Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        
        plt.figure(figsize=(10, 6))
        plt.contourf(xx, yy, Z, alpha=0.8)
        plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, edgecolors='k', marker='o', s=60)
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.title('SVM Decision Boundary and Test Data')
        plt.colorbar()
        plt.show()
    
    # If the data is not 2D, you can skip the decision boundary plot and just show a confusion matrix
    else:
        print("Data is not 2D, skipping decision boundary plot.")
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
class_names = ['edible', 'poisonous']  
visualize_classification_results(model=best_model, X_test=X_test, y_test=y_test, class_names=class_names)
```

# 1. Compare SVM Performance with Various Kernels (Linear, Polynomial, Radial Basis Function)

### Explanation:
- **Linear Kernel**: The simplest kernel, useful for linearly separable data. It performs well when the classes are already linearly separable or nearly so.
- **Polynomial Kernel**: Works well for cases where the data is not linearly separable, but a higher-dimensional feature space could make the data separable. However, it might be more prone to overfitting.
- **Radial Basis Function (RBF) Kernel**: This is the most commonly used kernel. It can capture complex relationships by mapping data into a higher-dimensional space, making it effective for datasets that are not linearly separable.

# 2. Analyze SVM Strengths and Weaknesses for the Mushroom Dataset Based on EDA and Visualization Results

### Strengths of SVM:
- **High Dimensionality**: SVM can perform well even in high-dimensional spaces, making it effective when dealing with large and complex datasets.
- **Effective with Non-Linear Data**: With the right kernel (like RBF), SVM can effectively handle non-linearly separable datasets, such as the Mushroom dataset, where the relationship between features and target might not be linear.
- **Clear Margin of Separation**: SVM aims to maximize the margin between classes, which generally leads to good generalization and low overfitting when the margin is large.

### Weaknesses of SVM:
- **Memory Intensive**: SVM requires a lot of memory, especially with large datasets. The training time increases significantly as the dataset size grows, making it inefficient for very large datasets.
- **Choice of Kernel**: Selecting the right kernel is crucial. If the data doesn't fit the kernel well (e.g., choosing a linear kernel for non-linear data), the performance can degrade.
- **Sensitivity to Noisy Data**: SVM is sensitive to noise and outliers. While it performs well in clean datasets, outliers can impact the margin and thus the model's accuracy.

### Visualization Results from EDA:
By plotting the feature importance or visualizing how the data is distributed, you can observe how well SVM is able to create a decision boundary, especially with kernels like RBF that can adapt to the data's complexities.

# 3. Discuss Practical Implications of SVM in Real-World Classification Tasks

SVM is highly regarded for its ability to produce robust classification results, especially in high-dimensional spaces and complex data. However, its real-world applicability depends on the dataset characteristics.

### Practical Implications:
- **Medical Diagnosis**: SVM is widely used in areas like medical diagnosis where separating classes with clear margins is important. For example, cancer cell detection, where the boundary between cancerous and non-cancerous cells is distinct, SVM can help with accurate classification.
- **Image Recognition**: SVM has been successful in image classification tasks, especially in situations where the image features can be extracted and represented as high-dimensional data.
- **Text Classification**: With high-dimensional feature sets such as word counts or TF-IDF scores, SVM can perform well in tasks like sentiment analysis or spam detection.

### Challenges in Real-World Use:
- **Scalability**: For large datasets, SVM can be computationally expensive, particularly during training. This makes it challenging for tasks with millions of data points.
- **Choice of Parameters**: In practical applications, hyperparameter tuning is crucial. SVM’s performance heavily depends on the kernel choice and parameters like `C` and `gamma`. It may require extensive cross-validation and grid search for optimal results.
- **Model Interpretability**: SVM models, especially with non-linear kernels, can be seen as black-box models. This might be an issue in industries like healthcare or finance, where model transparency is crucial for decision-making.

### Conclusion:
SVM is a powerful tool, but its performance depends on proper tuning and dataset characteristics. For the Mushroom dataset, SVM can perform well with the appropriate kernel (like RBF), but challenges include memory requirements and sensitivity to outliers. In real-world applications, SVM’s strengths in handling high-dimensional data are offset by scalability issues and the need for careful parameter tuning.
