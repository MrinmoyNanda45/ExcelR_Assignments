### Data visualization

### Bar plot

```python
def plot_bar_graphs(data, features, n_cols=3, figsize=(20, 30)):
    """
    Generate bar graphs for categorical features.

    Parameters:
        data (DataFrame): The input DataFrame containing categorical features.
        features (list): List of categorical features to plot.
        n_cols (int): Number of columns in the plot grid.
        figsize (tuple): Overall figure size for the plots.

    Returns:
        None
    """
    plt.style.use('seaborn-v0_8-darkgrid')
    n_features = len(features)
    n_rows = (n_features + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
    fig.suptitle('Bar Graphs of Categorical Features', fontsize=18, y=1.02)

    axes = axes.flatten()
    for idx, feature in enumerate(features):
        sns.countplot(data=data, x=feature, palette='viridis', ax=axes[idx], order=data[feature].value_counts().index)
        axes[idx].set_title(f'{feature} Bar Plot', fontsize=14)
        axes[idx].set_xlabel(feature, fontsize=12)
        axes[idx].set_ylabel('Count', fontsize=12)
        axes[idx].tick_params(axis='x', rotation=45)

    for idx in range(n_features, len(axes)):
        fig.delaxes(axes[idx])

    plt.tight_layout()
    plt.show()
```

```python
plot_bar_graphs(data= data, features= categorical_features)
```

### Stacked plot

```python
def plot_stacked_bar_graphs(data, features, target_feature, n_cols=3, figsize=(20, 35)):
    """
    Generate stacked bar graphs for multiple categorical features against a target feature.

    Parameters:
        data (DataFrame): The input DataFrame containing categorical features.
        features (list): List of categorical features to plot.
        target_feature (str): The target categorical feature for stacking.
        n_cols (int): Number of columns in the plot grid.
        figsize (tuple): Overall figure size for the plots.

    Returns:
        None
    """
    n_features = len(features)
    n_rows = (n_features + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
    fig.suptitle(f'Stacked Bar Graphs against {target_feature}', fontsize=18, y=1.02)

    axes = axes.flatten()
    for idx, feature in enumerate(features):
        stacked_data = data.groupby([feature, target_feature]).size().unstack(fill_value=0)
        stacked_data.plot(kind='bar', stacked=True, ax=axes[idx], colormap='viridis', alpha=0.8)
        axes[idx].set_title(f'{feature} vs {target_feature}', fontsize=14)
        axes[idx].set_xlabel(feature, fontsize=12)
        axes[idx].set_ylabel('Count', fontsize=12)

    for idx in range(n_features, len(axes)):
        fig.delaxes(axes[idx])

    plt.tight_layout()
    plt.show()
```

```python
plot_stacked_bar_graphs(data= data, features= categorical_features, target_feature= 'class')
```

### Heat map

```python
def plot_heatmaps(data, features, class_feature, n_cols=2, figsize=(20, 30)):
    """
    Generate heatmaps for categorical features with respect to a class variable.

    Parameters:
        data (DataFrame): The input DataFrame containing categorical features.
        features (list): List of categorical features to plot.
        class_feature (str): The class variable to compare against.
        n_cols (int): Number of columns in the plot grid.
        figsize (tuple): Overall figure size for the plots.

    Returns:
        None
    """
    plt.style.use('seaborn-v0_8-darkgrid')
    n_features = len(features)
    n_rows = (n_features + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
    fig.suptitle('Heatmaps of Categorical Features with Respect to Class', fontsize=18, y=1.02)

    axes = axes.flatten()
    for idx, feature in enumerate(features):
        heatmap_data = pd.crosstab(index=data[feature], columns=data[class_feature])
        sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='viridis', cbar=True, linewidths=0.5, ax=axes[idx])
        axes[idx].set_title(f'{feature} vs {class_feature}', fontsize=14)
        axes[idx].set_xlabel(class_feature, fontsize=12)
        axes[idx].set_ylabel(feature, fontsize=12)

    for idx in range(n_features, len(axes)):
        fig.delaxes(axes[idx])

    plt.tight_layout()
    plt.show()
```

```python
plot_heatmaps(data= data, features= categorical_features, class_feature= 'class')
```

### Pie Chart

```python
def plot_pie_charts(data, features, n_cols=3, figsize=(20, 30)):
    """
    Generate pie charts for multiple categorical features.

    Parameters:
        data (DataFrame): The input DataFrame containing categorical features.
        features (list): List of categorical features to plot.
        n_cols (int): Number of columns in the plot grid.
        figsize (tuple): Overall figure size for the plots.

    Returns:
        None
    """
    n_features = len(features)
    n_rows = (n_features + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
    fig.suptitle('Pie Charts of Categorical Features', fontsize=18, y=1.02)

    axes = axes.flatten()
    for idx, feature in enumerate(features):
        feature_counts = data[feature].value_counts()
        axes[idx].pie(feature_counts, labels=feature_counts.index, autopct='%1.1f%%', 
                      colors=sns.color_palette('viridis', len(feature_counts)), startangle=90)
        axes[idx].set_title(f'{feature} Pie Chart', fontsize=14)

    for idx in range(n_features, len(axes)):
        fig.delaxes(axes[idx])

    plt.tight_layout()
    plt.show()
```

```python
plot_pie_charts(data= data,features= categorical_features)
```

### Scatter Plot

```python
def plot_scatterplots(data, features, target, n_cols=2, figsize=(10, 5)):
    """
    Generate scatter plots for numerical features against a target variable.

    Parameters:
        data (DataFrame): The input DataFrame containing numerical features.
        features (list): List of numerical features to plot.
        target (str): Target variable for scatter plots.
        n_cols (int): Number of columns in the plot grid.
        figsize (tuple): Overall figure size for the plots.

    Returns:
        None
    """
    n_features = len(features)
    n_rows = (n_features + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
    fig.suptitle('Scatter Plots of Numerical Features Against Target', fontsize=18, y=1.02)

    axes = axes.flatten()
    for idx, feature in enumerate(features):
        sns.scatterplot(data=data, x=feature, y=target, ax=axes[idx], color='purple', alpha=0.7)
        axes[idx].set_title(f'{feature} vs {target}', fontsize=14)
        axes[idx].set_xlabel(feature, fontsize=12)
        axes[idx].set_ylabel(target, fontsize=12)

    for idx in range(n_features, len(axes)):
        fig.delaxes(axes[idx])

    plt.tight_layout()
    plt.show()
```

### Encoding

- Ohe will not be fessible here as here we have already 24 categorical columns
- so we will use target guided encoding
- the probability of each class (i.e., the proportion of observations of each class within a category of the feature) as the encoded value.

```python
def calculate_target_probabilities(df, categorical_columns, target_column):
    """
    Calculate the target probabilities for each category in each categorical feature.

    This function computes the conditional probabilities of the target column for each category 
    of the specified categorical features in the DataFrame. The probabilities are normalized 
    counts of the target classes per category.

    Args:
        df (pd.DataFrame): The input DataFrame containing the data.
        categorical_columns (list): List of categorical column names to encode.
        target_column (str): The target column name whose class probabilities need to be calculated.

    Returns:
        dict: A dictionary where the keys are categorical column names and the values are dictionaries 
              containing the class probabilities for each category.
    """
    probabilities = {}
    for column in categorical_columns:
        probabilities[column] = {}
        category_counts = df.groupby(column)[target_column].value_counts(normalize=True).unstack()
        for category in category_counts.index:
            probabilities[column][category] = category_counts.loc[category].to_dict()
    return probabilities
```

```python
def target_encode_feature(df, feature_column, target_column, probabilities):
    """
    Perform target encoding for a single categorical feature based on the target column.

    This function encodes a categorical feature by replacing each category with the target class 
    probabilities. It maps each category of the feature to the probability of the target classes.

    Args:
        df (pd.DataFrame): The input DataFrame containing the data.
        feature_column (str): The categorical feature column to be encoded.
        target_column (str): The target column name used to calculate the probabilities.
        probabilities (dict): A dictionary of class probabilities for each category in each categorical feature.

    Returns:
        pd.DataFrame: A DataFrame containing the encoded features with columns representing the 
                      class probabilities for each unique class of the target column.
    """
    encoded_column = df[feature_column].map(lambda x: probabilities[feature_column].get(x, {}))
    encoded_df = pd.DataFrame(encoded_column.tolist())
    columns = [f"{feature_column}_{target_column}_{cls}" for cls in df[target_column].unique()]
    encoded_df.columns = columns
    return encoded_df
```

```python
def target_encode_all_features(df, categorical_columns, target_column):
    """
    Perform target encoding for all specified categorical features in the DataFrame.

    This function applies target encoding to all categorical features by calculating the 
    target probabilities and encoding each feature based on the probabilities for the 
    target column. The encoded features are concatenated into a final DataFrame.

    Args:
        df (pd.DataFrame): The input DataFrame containing the data.
        categorical_columns (list): List of categorical column names to encode.
        target_column (str): The target column name used to calculate the probabilities.

    Returns:
        pd.DataFrame: A DataFrame with the target-encoded features concatenated.
        dict: A dictionary containing the class probabilities for each category in each categorical feature.
    """
    probabilities = calculate_target_probabilities(df, categorical_columns, target_column)
    encoded_dfs = []
    
    for column in categorical_columns:
        encoded_df = target_encode_feature(df, column, target_column, probabilities)
        encoded_dfs.append(encoded_df)
    
    final_encoded_df = pd.concat(encoded_dfs, axis=1)
    return final_encoded_df, probabilities
```
