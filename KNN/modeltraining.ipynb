### Scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
def scale_standard_train_test(X_train, X_test):
    """
    Scales the numerical features of the training and test sets using Standard scaling.
    The scaling is done only on the training data to avoid data leakage.

    Parameters:
        data (pd.DataFrame): The dataset containing the numerical features.
        features (list): A list of numerical features to scale.

    Returns:
        X_train_scaled (pd.DataFrame): Scaled training features.
        X_test_scaled (pd.DataFrame): Scaled testing features.
    """
    scaler = StandardScaler()
    
    # Fit on the training data and transform the training data
    X_train_scaled = scaler.fit_transform(X_train)
    
    # Only transform the test data using the scaler fitted on the training data
    X_test_scaled = scaler.transform(X_test)
    
    return X_train_scaled, X_test_scaled
```

```python
X_train_scaled, X_test_scaled = scale_standard_train_test(X_train= X_train, X_test= X_test)
```

### Model Training

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def train_knn_classifier(X_train, y_train, n_neighbors=5, metric='minkowski'):
    """
    Trains the KNN classifier on the training data.

    Parameters:
        X_train (ndarray or pd.DataFrame): Training features.
        y_train (ndarray or pd.Series): Training target values.
        n_neighbors (int): Number of neighbors to consider for KNN. Default is 5.
        metric (str): Distance metric to use. Default is 'minkowski'.

    Returns:
        KNeighborsClassifier: Fitted KNN classifier.
    """
    knn = KNeighborsClassifier(n_neighbors=n_neighbors, metric=metric)
    knn.fit(X_train, y_train)
    return knn
```

```python
knn_model = train_knn_classifier(X_train= X_train, y_train= y_train)
```

### Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
def evaluate_classifier(y_test, y_pred):
    """
    Evaluates the classifier's performance using accuracy, precision, recall, and F1-score.

    Parameters:
        y_test (ndarray or pd.Series): Actual target values from the test set.
        y_pred (ndarray or pd.Series): Predicted target values from the classifier.

    Returns:
        dict: Dictionary containing accuracy, precision, recall, and F1-score.
    """
    evaluation_metrics = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, average='weighted'),
        'Recall': recall_score(y_test, y_pred, average='weighted'),
        'F1 Score': f1_score(y_test, y_pred, average='weighted')
    }
    return evaluation_metrics
```

```python
metrics = evaluate_classifier(y_test,y_pred)
```

### Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
def random_search_knn(X_train, y_train):
    """
    Perform Random Search to find the best hyperparameters for KNN.

    Parameters:
        X_train (ndarray or pd.DataFrame): Training features.
        y_train (ndarray or pd.Series): Training target values.

    Returns:
        best_knn (KNeighborsClassifier): Best KNN model after random search.
        best_params (dict): Best hyperparameters found.
    """
    param_dist = {
        'n_neighbors': np.random.randint(1, 21, 20),  
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan'],
        'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    }

    knn = KNeighborsClassifier()
    random_search = RandomizedSearchCV(knn, param_dist, n_iter=100, cv=5, scoring='accuracy', n_jobs=-1, random_state=42)

    random_search.fit(X_train, y_train)

    best_knn = random_search.best_estimator_
    best_params = random_search.best_params_

    print(f"Best hyperparameters: {best_params}")
    print("Best model performance on training data:")
    print(classification_report(y_train, best_knn.predict(X_train)))

    return best_knn, best_params
```

```python
best_knn_model , best_params = random_search_knn(X_train= X_train, y_train= y_train)
```

```python
print("Best model performance on testing data:")
print(classification_report(y_test, best_knn_model.predict(X_test)))
```

### Decision Boundary

```python
from itertools import combinations
```

```python
def plot_pairwise_decision_boundaries(X, y, model, feature_names, figsize=(10, 8)):
    """
    Visualizes decision boundaries for all feature pairs of the dataset. 
    Each plot shows how the KNN classifier divides the feature space using two features at a time.
    """

    feature_pairs = list(combinations(range(X.shape[1]), 2))
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    for idx, (i, j) in enumerate(feature_pairs):
        plt.figure(figsize=figsize)
        
        X_pair = X_scaled[:, [i, j]]
        model.fit(X_pair, y)

        h = .02
        x_min, x_max = X_pair[:, 0].min() - 1, X_pair[:, 0].max() + 1
        y_min, y_max = X_pair[:, 1].min() - 1, X_pair[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                             np.arange(y_min, y_max, h))

        Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)

        ax = plt.gca()  # Get the current axes
        ax.contourf(xx, yy, Z, alpha=0.8, cmap='coolwarm')
        scatter = ax.scatter(X_pair[:, 0], X_pair[:, 1], c=y, edgecolors='k', cmap='coolwarm', s=50)

        ax.set_title(f'{feature_names[i]} vs {feature_names[j]}', fontsize=14, weight='bold')
        ax.set_xlabel(feature_names[i], fontsize=12)
        ax.set_ylabel(feature_names[j], fontsize=12)

        ax.tick_params(axis='both', labelsize=10)
        ax.spines['top'].set_visible(False)  # Modify spines through the axes object
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_linewidth(0.5)
        ax.spines['bottom'].set_linewidth(0.5)

        ax.legend(*scatter.legend_elements(), title="Classes", loc="upper right", fontsize=10)

        plt.tight_layout()
        plt.show()
```

```python
plot_pairwise_decision_boundaries(X= X, 
                                  y= y, 
                                  model= best_knn_model, 
                                  feature_names= X.columns.to_list())
```
